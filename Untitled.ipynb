First implement lambda cdm cosmology, expressed as

$$ H(z) = H_0 \sqrt{\Omega_m\cdot(1+z)^3 + \Omega_\Lambda + \Omega_k \cdot (1+z)^2} $$,

where $\Omega_\Lambda = 1-\Omega_m$, to reduce the number of parameters to be evaluated. $H_0$ is calibrated to late-time measurements from BOSS.

# Functions

```python
# ---- Constants ----

# speed of light in km / s
c = 299792.458

# Sound horizon used in fiducial cosmology (BOSS)
r_fid = 147.78
```

```python
# TO DO: Add wCDM model

def H_LambdaCDM(z, H_0, O_m, O_k):
    
    return H_0 * (O_m * (1+z) ** 3 + (1-O_m) + O_k * (1+z) ** 2) ** 0.5

def ComovingDistance(H_inv, H_0, z, O_k):
    
    DH = c / H_0
    Dm = cumtrapz(H_inv, z, initial=0) * c
    
    if O_k > 0.0:
        
        Dm = DH / O_k ** 0.5 * np.sinh( O_k ** 0.5 * Dm / DH)
    
    elif O_k < 0.0:
        
        Dm = DH / np.abs(O_k) ** 0.5 * np.sin( np.abs(O_k) ** 0.5 * Dm / DH)
    
    return Dm

def TimeDelayDistance(z_l, z_s, Dm_array, z_array, O_k, H_0):
    
    Da_array = Dm_array / (1 + z_l)
    Da_l = Da_array[z_array == z_l]
    Da_s = Da_array[z_array == z_s]
    Dm_l = Dm_array[z_array == z_l]
    Dm_s = Dm_array[z_array == z_s]
    
    Da_ls = 1 / (1 + z_l) * ( Dm_s * np.sqrt(1 + O_k * ) )
    
    ddt = (1+z_l) * Da_l * Da_s / Da_ls

def Compute_LLH(r, sigma):
    
    if sigma.ndim == 1:
        LLH = -0.5 * np.sum(r ** 2 / sigma **2)
    
    else:
        LLH = -0.5 * r.T @ sigma @ r
    
    return LLH
```

# Prior and likelihood

```python
def Ln_Prior(theta):
    
    H_0, O_m, O_k, r_s, M = theta
    
    # H_0
    if not 25 < H_0 < 100:
        return -np.inf
    
    # O_m
    if not 0.05 < O_m < 0.75:
        return -np.inf
    
    # O_k
    if not -1 < O_k < 1:
        return -np.inf
    
    # r_s
    if not 0 < r_s < 200:
        return -np.inf
    
    return 0.0

def Ln_Likelihood(theta):
    
    global z_pan
    global mu_pan
    global invcov_pan
    global z_BAO
    global Dm_BAO
    global H_BAO
    global invcov_BAO
    
    # Unpack parameters
    H_0, O_m, O_k, r_s, M = theta
    
    # Combine redshifts
    z_array = np.linspace(0, 1.8, 1000)
    z_total = np.concatenate((z_pan, z_BAO, z_array))
    z_total.sort()
    
    # Obtain H, Dm and Dl
    H = H_LambdaCDM(z_total, H_0, O_m, O_k)
    Dm = ComovingDistance(1. / H, H_0, z_total, O_k)
    Dl = (1+z_total) * Dm
    
    # ---------------------------------

    # - - - - - SH0ES - - - - -
    
    r_SH0ES = H_0 - 74.03  # Â± 1.42
    LLH_SH0ES = - 0.5 * r_SH0ES ** 2 / 1.42 ** 2
    
    # - - - - - SN - - - - -
    
    # Calculate distance modulus
    SN_Dl = Dl[np.in1d(z_total, z_pan)]
    SN_mu = M + 5 * np.log10(SN_Dl)
    
    # Compute llh for SN
    r_SN = mu_pan - SN_mu
    LLH_SN = Compute_LLH(r_SN, invcov_pan)
    
    # - - - - - BAO - - - - -
    
    # Get H for BAO
    BAO_H = H[np.in1d(z_total, z_BAO)]
    H_fiducial = BAO_H * r_s / r_fid
    
    # Get Dm for BAO
    BAO_Dm = Dm[np.in1d(z_total, z_BAO)]
    Dm_fiducial = BAO_Dm * r_fid / r_s
    
    # Compute llh for BAO
    r_BAO = np.concatenate((Dm_BAO-Dm_fiducial, H_BAO - H_fiducial))
    LLH_BAO = Compute_LLH(r_BAO, invcov_BAO)
    
    # - - - - - - - - - - - - - - - - -
    
    total_LLH = LLH_SH0ES + LLH_SN + LLH_BAO
    
    if np.isnan(total_LLH):
        return -np.inf
    
    return total_LLH

def Ln_Posterior(theta):
    
    prior = Ln_Prior(theta)
    likelihood = Ln_Likelihood(theta)

    return prior + likelihood
```

```python
nwalkers = 100

 # For theta: Define a random starting position for each of the walkers [nwalkers, ndim].
M_init = np.random.uniform(0, 30, nwalkers)
r_s_init = np.random.uniform(100, 150, nwalkers)
O_k_init = np.random.uniform(-0.45, 0.45, nwalkers)
H_0_init = np.random.uniform(25, 100, nwalkers)
O_m_init = np.random.uniform(0.05, 0.5, nwalkers)

theta_init = np.array([H_0_init, O_m_init, O_k_init, r_s_init, M_init]).T

ndims = theta.shape[1]
```

```python
Nburnin = 5000
Nsamples = 5000

sampler = em.EnsembleSampler(nwalkers, ndims, Ln_Posterior)
```

```python
sampler.run_mcmc(theta_init, Nsamples+Nburnin);
```

```python
acl = max(sampler.get_autocorr_time(quiet=True))
samples = sampler.chain[:, Nburnin::int(acl), :].reshape((-1, ndims))
```
